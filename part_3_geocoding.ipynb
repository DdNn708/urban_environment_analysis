```python
df1 = df[(-df.lon.isnull()) & (df.raw_class != 'highway')]
print(len(df1))

m = folium.Map(location=[df1.lat.mean(), df1.lon.mean()],
              zoom_start=12, weight=1000, height=500,
              tiles='cartodbpositron')

for lat, lon in zip(df1.lat, df1.lon):
        folium.Circle(
        location=[lat, lon],
        radius=60,
        stroke=False,
        fill=True,
        fill_color='purple',
        fill_opacity=0.1
    ).add_to(m)
        
m
```
